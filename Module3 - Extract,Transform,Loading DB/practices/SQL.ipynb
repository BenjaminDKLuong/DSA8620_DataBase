#### M3:P2:Q1: Load Data for Albums

Using Python load the Albums into the database 


````python
%%sql 
DROP TABLE IF EXISTS Album;
DROP TABLE IF EXISTS Artist;
DROP TABLE IF EXISTS Song;

CREATE TABLE Artist (
    artist_id INT, 
    name varchar(100), 
    genre varchar(100), 
    year_formed INT,
    PRIMARY KEY (artist_id)
);


````

````python
%%sql 
CREATE TABLE Album (
    album_id INT,
    artist_id INT, 
    title varchar(100), 
    year_produced INT,
    PRIMARY KEY (album_id),
    FOREIGN KEY (artist_id)
        REFERENCES Artist(artist_id)
);

CREATE TABLE Song (
    title varchar(100),
    track INT,
    length varchar(20),
    FOREIGN KEY (title)
        REFERENCES Album(title)

);
````

````python
%sql SELECT * FROM Album;
````

````python
# Write your Python here
# load database
# ---------------------
import pandas  # requires the xlrd optional library for Pandas
df_albums = pandas.read_excel(open('../../../datasets/Module4Data.xlsx','rb'), sheetname='Albums',header= None)
df_albums
````

````python
df_albums.info()
````

````python
# write data into database
# ---------------------
import sqlite3
import numpy

databaseFilename = '../databases/songs.db'
connection = sqlite3.connect(databaseFilename)
cursor = connection.cursor()

sqlite3.register_adapter(numpy.int64, int)
sqlite3.register_adapter(numpy.float64, float)

for row in df_albums.itertuples(index=True, name ='None'):
    print(row)
    cursor.execute('INSERT INTO Album VALUES(?,?,?,?)',row)

# Save (commit) the changes
connection.commit()

````

#### M3:P2:Q3: How many bands are in the database?

````python

%%sql 
SELECT DISTINCT artist_id FROM Album ;
````

````python
%%sql 
SELECT count(*) as artist_number FROM (SELECT DISTINCT artist_id FROM Album);
````

#### M3:P2:Q4: How many songs are played by each band?

````python

%%sql 

SELECT artist_id, title , count(track)FROM Album
JOIN (Song) USING (title)
GROUP BY title
ORDER BY artist_id;



````

````python

%%sql 
SELECT artist_id, sum(number_track)
FROM(
    SELECT artist_id, title , count(track) as number_track FROM Album
    JOIN (Song) USING (title)
    GROUP BY title
    ORDER BY artist_id)
GROUP BY artist_id;



````

#### M3:P2:Q5: What is the average song length?

````python

%%sql 
SELECT AVG(length) FROM Song;
````

#### M3:P2:Q6: What is the average song length for The Beatles, per album?

````python
%%sql 
SELECT * FROM
(SELECT artist_id,title , AVG(length)
FROM Album
JOIN (Song) USING (title)
GROUP BY title
ORDER BY artist_id)
WHERE artist_id = 'The Beatles'
;
````
